 What is the role of cellulose during the colonization of the ALI by *Pseudomonas fluorescens SBW25*

### _Motivation :

Macroscopic observations of static microcosms inoculated with *Pseudomonas fluorescens SBW25* have shown that collective groups form at the Air Liquid Interphase (ALI) upon bacterial growth. 

The dynamics and emergence of collectives and the formation of groups is of interest for us to understand how cellulose helps biofilm growth and establishment of bacteria at the ALI. 

Here we study the initiation of these collectives prior biofilm establishment at the ALI and delve into the role of cellulose as biofilm initiator/facilitator. We use a device that serves the purpose to expose a triple point contact where bacteria attaches, the liquid media and air and quantify the interactions between bacterial secreted cellulose and the triple contact point. 

We utilize the Smooth Type wild type (SM) able to produce cellulose and form biofilms and the dwss or cellulose deficient mutant, whose cellulose operon WSS has been deleted from the genome.
